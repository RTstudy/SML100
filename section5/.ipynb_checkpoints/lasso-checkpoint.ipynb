```python
# define soft
def soft_th(lam, x):
    return np.sign(x)*np.maximum(np.abs(x)-lam, 0)
```

```python
# define lasso
def lasso(x,y,lam=0):
    X = copy.copy(x)  # もとのオブジェクトが影響を受けないようにcopyを使用
    n,p = X.shape     # 説明変数の行列サイズを算出
    
    # Xを標準化
    # 本文ではforループを使っているがaxisで平均、標準偏差をどの方向で実施するかをaxis引数で指定すれば良い
    X_bar = np.mean(X, axis = 0)  # 平均を算出
    s = np.std(X, axis = 0)      # 標準偏差を算出
    X = (X - X_bar) / s
    
    # 切片を0にする(=yを中心化)
    y_bar = np.mean(y)
    y = y - y_bar
    
    # 収束条件用の変数を初期化
    eps = 1
    
    # betaを初期化
    beta = np.zeros(p); beta_old = np.zeros(p)
    
    # betaを算出するループ
    while eps > 0.0001:
        for j in range(p):
            index = list(set(range(p))-{j})  # betaのセットリストを作成し、jのみ除く
            r = y - X[:, index]@beta[index]  # indexに入っているbetaで計算された値をyから除く = のぞかれた変数で説明されるyの部分のみを残して従属変数とする
            beta[j] = soft_th(lam, r.T@X[:, j]/n)
        eps = np.max(np.abs(beta-beta_old))
        beta_old = beta
    
    # 標準化したデータでbetaを算出していたので標準偏差で割り戻す
    for j in range(p):
        beta[j] = beta[j]/s[j]
        
    beta_O = y_bar - X_bar.T@beta
    return {'beta':beta, 'beta_O':beta_O}
```

```python
lasso(X, y, 20)
```

```python
lambda_seq = np.arange(0,200,0.5)
plt.xlim(0, 200)
plt.ylim(-10, 20)
plt.xlabel('lambda')
plt.ylabel('beta')
labels = ['警察への年間資金',
          '25歳以上で高校を卒業した人の割合',
          '16-19歳で高校に通っていない人の割合',
          '18-24歳で大学生の割合',
          '25歳以上で4年制大学を卒業した人の割合']
for j in range(p):
    coef_seq = []
    for l in lambda_seq:
        coef_seq.append(lasso(X, y, l)['beta'][j])
    plt.plot(lambda_seq, coef_seq, label='{}'.format(labels[j]))
plt.legend(loc='upper right')
plt.title('各λについての各係数の値')
```
